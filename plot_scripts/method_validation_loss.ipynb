```python
from os import listdir
from os.path import isfile, join


# path = r"/Users/paopao_ch/Documents/projects/v1_modelling/Mouse-V1-Pytorch/results/backprop1000/method_validation"
# path = r"/Users/paopao_ch/Documents/projects/v1_modelling/Mouse-V1-Pytorch/results/uniform_contrast_value/method_validation"
# path = r"/Users/paopao_ch/Documents/projects/v1_modelling/Mouse-V1-Pytorch/results/non_uniform_contrast_high_contrast/method_validation"
path = r"/Users/paopao_ch/Documents/projects/v1_modelling/Mouse-V1-Pytorch/results/backprop1000NewLossFunctionHeterFF/method_validation"
# path = r"/Users/paopao_ch/Documents/projects/v1_modelling/results/actual_data/method_validation_actual_data_gd"
files = [f for f in listdir(path) if isfile(join(path, f))]
```

```python
losses = []

for filename in files:
    loss = []
    with open(f"{path}/{filename}", "r") as f:
        for line in f:
            words = line.split(" ")
            if words[0] == "loss:":
                loss.append(float(words[1]))
    losses.append(loss)
```

```python
for loss in losses:
    plt.plot(loss)
plt.ylim(0.8, 2)
plt.show()
```
