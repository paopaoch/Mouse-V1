```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from os import listdir
from os.path import isfile, join

path = r"/Users/paopao_ch/Documents/projects/v1_modelling/Mouse-V1-Pytorch/results/contrast_varying"
files = [f for f in listdir(path) if isfile(join(path, f))]
files.sort()
files.pop(-1)
files
```

```python
losses = []

for filename in files:
    loss = []
    with open(f"{path}/{filename}", "r") as f:
        for line in f:
            words = line.split(" ")
            if words[0] == "loss:":
                loss.append(float(words[1]))
    losses.append(loss)
```

```python
contrast_types = ["Low contrast skew", "Uniform", "High contrast skew"]

for loss, contrast_type in zip(losses, contrast_types):
    plt.plot(loss, label=f"Sample size {contrast_type}")
plt.ylim(0, 0.25)
plt.legend()
plt.show()
```
