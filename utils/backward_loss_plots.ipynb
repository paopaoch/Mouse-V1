```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import os
```

```python

file_name = r"/Users/paopao_ch/Documents/projects/v1_modelling/Mouse-V1-Pytorch/log_backwards.log"

losses = {
    "loss" : [],
}

with open(file_name, "r") as f:
    for line in f:
        words = line.split(" ")
        if words[0] == "loss:":
            losses["loss"].append(float(words[1]))
# df = pd.DataFrame(data=losses)
```
