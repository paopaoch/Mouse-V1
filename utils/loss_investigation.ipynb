```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
loss1 = pd.read_csv("../data/MMD_1_init.csv", header=None).to_numpy().squeeze()
loss2 = pd.read_csv("../data/MMD_2_init.csv", header=None).to_numpy().squeeze()
```

```python
def print_loss_analysis(loss, bins=None):
    print("length:", len(loss))
    print("mean:", loss.mean())
    print("std:", loss.std())
    print("CV:", loss.std() / loss.mean())
    print("max:", loss.max())
    print("min:", loss.min())
    print("range:", loss.max() - loss.min())
    if bins is None:
        plt.hist(loss)
    else:
        plt.hist(loss, bins=bins)
    plt.show()
```

```python
bin_size = 0.00005
bins = np.arange(0.0955, 0.096 + bin_size, bin_size)
print_loss_analysis(loss1, bins)
```

```python
bin_size = 0.00005
bins = np.arange(0.0063, 0.00665 + bin_size, bin_size)
print_loss_analysis(loss2, bins)
```

## Try average across trials

```python
def average_of_i(arr, num_avg=2):
    result = []
    for i in range(0, len(arr), num_avg):
        slice_ = arr[i:i+num_avg]
        avg = sum(slice_) / len(slice_)
        result.append(avg)
    return np.array(result)
```

```python
print_loss_analysis(average_of_i(loss1))
print_loss_analysis(average_of_i(loss1, num_avg=3))
print_loss_analysis(average_of_i(loss1, num_avg=4))
print_loss_analysis(average_of_i(loss1, num_avg=5))
```

```python
print_loss_analysis(average_of_i(loss2))
print_loss_analysis(average_of_i(loss2, num_avg=3))
print_loss_analysis(average_of_i(loss2, num_avg=4))
print_loss_analysis(average_of_i(loss2, num_avg=5))
```
