```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

file_name = r"/Users/paopao_ch/Documents/projects/v1_modelling/Mouse-V1-Pytorch/results/backprop1000/method_validation"

losses = []

for i in range(17):
    loss = []
    with open(f"{file_name}/loss{i+1}.log", "r") as f:
        for line in f:
            words = line.split(" ")
            if words[0] == "loss:":
                loss.append(float(words[1]))
    losses.append(loss)
```

```python
for loss in losses:
    plt.plot(loss)
plt.ylim(0, 1)
plt.show()
```

```python
for i, loss in enumerate(losses):
    plt.plot(loss, label=f"{i}")
plt.ylim(0.005, 0.02)
plt.xlim(0, 70)
plt.legend()
plt.show()
```
